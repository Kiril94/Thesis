```python
rgb = plt.imread(r"cat.jpg")
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.144])
img = rgb2gray(rgb)
fig, ax = plt.subplots()
ax.imshow(img, cmap='Greys')
```

```python
## Functions to go from image to frequency-image and back
im2freq = lambda data: fftpack.rfft(fftpack.rfft(data, axis=0),
                               axis=1)
freq2im = lambda f: fftpack.irfft(fftpack.irfft(f, axis=1),
                             axis=0)
def normalize(img):
    img_max = np.max(img)
    img_min = np.min(img)
    return (img-img_min)/(img_max-img_min)
    
```

```python
freq_img = freq2im(img)
norm_freq = normalize(freq_img)
plt.imshow(norm_freq[20:30,0:10])
```
