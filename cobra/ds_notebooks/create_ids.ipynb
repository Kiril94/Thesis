```python
import pandas as pd
import os
from os.path import split, join
import feather
cobra_dir = split(os.getcwd())[0]
tables_dir = join(cobra_dir, 'data', 'tables')
```

```python
df_all = pd.read_csv(join(tables_dir, 'neg_pos.csv'))
dfc = pd.read_csv(join(tables_dir, 'neg_pos_clean.csv'))
df_all.head()
```

```python
print('There are ', len(df_all), 'MRIs')
print('We need therefore an index of length 6')
df_all['newID'] = df_all.index.map(lambda x: str(x).zfill(6))
df_all.newID
```

```python
dfc_new = dfc.merge(df_all[['SeriesInstanceUID', 'newID']], on='SeriesInstanceUID', how='left') 
```

```python
df_all.to_csv(join(tables_dir, 'neg_pos.csv'), index=False, header=True)
dfc_new.to_csv(join(tables_dir, 'neg_pos_clean.csv'), index=False, header=True)
```

```python
feather.write_dataframe(df_all,join(tables_dir, 'neg_pos.feather'))
feather.write_dataframe(dfc_new,join(tables_dir, 'neg_pos_clean.feather'))
```
