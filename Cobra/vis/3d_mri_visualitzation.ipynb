```python
def explore_3d_image(layer):
    plt.figure(figsize=(10,15))
    plt.imshow(image_data[:,:,layer],cmap='gray',aspect='auto')
    plt.axis('off')
    return layer

#run the interact function to explore the data
interact(explore_3d_image,layer=(0,image_data.shape[2]-1))
    
```
